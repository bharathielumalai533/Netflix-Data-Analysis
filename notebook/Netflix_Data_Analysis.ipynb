```python
df['director'] = df['director'].fillna('Unknown')
```

```python
df['cast'] = df['cast'].fillna('Not Available')
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df = df.dropna(subset=['date_added', 'rating', 'duration'])
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/netflix_cleaned.csv", index=False)
```

```python
df['date_added'] = df['date_added'].str.strip()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
```

```python
top_country = df['country'].value_counts().head(10)

top_country
```

```python
release_year = df['release_year'].value_counts().sort_index()

release_year
```

```python
plt.figure(figsize=(14,6))

release_year.plot(kind='line', marker='o')

plt.title("Netflix Content Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.grid(True)

plt.show()
```

```python
top_directors = (
    df[df['director'] != 'Unknown']['director']
    .value_counts()
    .head(10)
)

top_directors
```

```python
plt.figure(figsize=(12,6))

top_directors.plot(kind='barh')

plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")

plt.show()
```

```python
# Split genres into individual rows
genres = df['listed_in'].str.split(',').explode().str.strip()

# Top 10 genres
top_genres = genres.value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
top_genres.plot(kind='bar', color='teal')

plt.title("Top 10 Netflix Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.show()
```

```python
rating = df['rating'].value_counts()

plt.figure(figsize=(10,6))
rating.plot(kind='bar', color='orange')

plt.title("Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")

plt.show()
```

```python
added_year = df['date_added'].dt.year.value_counts().sort_index()

plt.figure(figsize=(12,6))
added_year.plot(kind='line', marker='o')

plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)

plt.show()
```

```python
month = df['date_added'].dt.month_name()

month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

month = month.value_counts().reindex(month_order)

plt.figure(figsize=(12,6))
month.plot(kind='bar', color='green')

plt.title("Netflix Content Added by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.replace(' min','').astype(int)

plt.figure(figsize=(12,6))
plt.hist(movies['duration'], bins=30)

plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")

plt.show()
```

```python
actors = df['cast'].str.split(',').explode().str.strip()

top_actors = actors.value_counts().head(10)

plt.figure(figsize=(12,6))
top_actors.plot(kind='barh', color='purple')

plt.title("Top 10 Actors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Actor")

plt.show()
```

```python
content_type = df['type'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    content_type,
    labels=content_type.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Movies vs TV Shows")

plt.show()
```

```python
top_year = df['release_year'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_year.plot(kind='bar', color='red')

plt.title("Top 10 Release Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
movie_rating = df[df['type']=='Movie']['rating'].value_counts()

plt.figure(figsize=(10,6))
movie_rating.plot(kind='bar')

plt.title("Movie Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Movies")

plt.show()
```
